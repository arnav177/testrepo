```python
import pandas as pd
import numpy as np
import seaborn as sns
import folium
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv')
```

# 1.1 Create a line plot Automobile sales during Recession

```python
df_line= df.groupby(['Year'])['Automobile_Sales'].mean()
                    
plt.figure(figsize=(10, 6))

df_line.plot(kind = 'line')
plt.xticks(list(range(1980,2024)), rotation = 75)
plt.xlabel('Year')
plt.ylabel('Automotive Sales')
plt.title('Automobile Sales during Recession')
plt.text(1982, 650, '1981-82 Recession')
plt.text(2008,500 , '2008-10 Recession')
plt.legend()
plt.show()
```

# 1.2 Vehicle Type wise sales During Recessions

```python
df_line= df.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()
df_line.set_index('Year', inplace=True)



df_line = df_line.groupby(['Vehicle_Type'])['Automobile_Sales']                     
plt.figure(figsize=(10, 6))
df_line.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Vehicle Sales')
plt.title('Sales Trend vehicle_wise during recession')
plt.legend()
plt.savefig('Line_Plot_1.png')
plt.show()
```

# 1.4 comparision of Recssion and Non - Recession  Period

```python
    rec_data = df[df['Recession'] == 1]
    non_rec_data = df[df['Recession'] == 0]
    
    #Figure
    fig=plt.figure(figsize=(12, 6))
    
    #Create different axes for subploting
    ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
    ax1 = fig.add_subplot(1 ,2 ,2 ) # add subplot 2 (1 row, 2 columns, second plot). 
    
    #plt.subplot(1, 2, 1)
    sns.lineplot(x='Year', y='GDP', data=rec_data, label='Recession', ax=ax0)
    ax0.set_xlabel('Year')
    ax0.set_ylabel('GDP')
    ax0.set_title('GDP Variation during Recession Period')
    
    #plt.subplot(1, 2, 2)
    sns.lineplot(x='Year', y='GDP', data=non_rec_data, label='Recession',ax=ax1)
    ax1.set_xlabel('Year')
    ax1.set_ylabel('GDP')
    ax1.set_title('GDP Variation during Non Recession Period')
    
    plt.tight_layout()
    plt.show()
```

# 1.7 Pie chart to display the portion of advertising expenditure of Automotives during recession and non-recession periods

```python
    data = df[df['Recession'] == 1]
    Data = df[df['Recession'] == 0]

    # Calculate the total advertising expenditure for both periods
    total = data['Advertising_Expenditure'].sum()
    Total = Data['Advertising_Expenditure'].sum()

    # Create a pie chart for the advertising expenditure 
    plt.figure(figsize=(8, 6))

    labels = ['Recession', 'Non-Recession']
    sizes = [total, Total]
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

    plt.title('Advertising Expenditure of Companies')

    plt.show()
```

# 1.8 Pie chart to display the total Advertisement expenditure for each vehicle type during recession

```python
    data = df[df['Recession'] == 1]

    # Calculate the sales volume by vehicle type during recessions
    sales = data.groupby('Vehicle_Type')['Automobile_Sales'].sum()

    # Create a pie chart for the share of each vehicle type in total sales during recessions
    plt.figure(figsize=(10, 6))

    labels = sales.index
    sizes = sales.values
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

    plt.title('Expenditure of Company During Recession')

    plt.show()
```

# 1.9 Develop a countplot to analyse the effect of the unemployment rate on vehicle type and sales during recession period

```python
df_rec = df[df['Recession']==1]
sns.lineplot(data=df_rec, x='unemployment_rate', y='Automobile_Sales',
             hue='Vehicle_Type', style='Vehicle_Type', markers='o', err_style=None)
plt.ylim(0,850)
plt.legend(loc=(0.05,.3))
plt.xlabel('unemployment Rate')
plt.ylabel('Automobile_Sales')
plt.title('Effect of the unemployment rate on vehicle type and sales during recessions')
plt.show()
```
